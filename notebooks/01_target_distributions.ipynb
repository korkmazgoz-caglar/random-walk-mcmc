### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

from rwmcmc.targets import gaussian_1d_log_pdf
```

### Creating the dataset

```python
x = np.linspace(-4, 4, 1000)
log_pdf = gaussian_1d_log_pdf(x, mu=0.0, sigma=1.0)
unnormalized_pdf = np.exp(log_pdf)
normalized_pdf = norm.pdf(x) # Normalized PDF for comparison, special function from scipy.stats

print(f"log_pdf: {log_pdf.max():.4f})\n")
print(f"unnormalized_pdf: {unnormalized_pdf.max():.4f})\n")
print(f"normalized_pdf: {normalized_pdf.max():.4f})\n")
```

### Plotting

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].plot(x, log_pdf, label="Log PDF (Unnormalized)", color="blue")
axes[0].set_title("Log of Unnormalized PDF")
axes[0].set_xlabel("x")
axes[0].set_ylabel("Log PDF")
axes[0].legend()
axes[0].grid(True)

axes[1].plot(x, unnormalized_pdf, label="Unnormalized PDF", color="red")
axes[1].set_title("Unnormalized PDF")
axes[1].set_xlabel("x")
axes[1].set_ylabel("PDF")
axes[1].legend()
axes[1].grid(True)

axes[1].plot(x, normalized_pdf, label="Normalized PDF", color="blue", linestyle="--")
axes[1].set_title("Normalized PDF")
axes[1].set_xlabel("x")
axes[1].set_ylabel("PDF")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
